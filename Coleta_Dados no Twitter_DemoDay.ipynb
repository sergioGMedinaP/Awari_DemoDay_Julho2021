 <h2 style ="color:green">(1) Remoção de hashtags, links, menções (@) </h2>

```python
import re

def limpando_chars(text):
    text = re.sub('@[A-Za-z0–9]+', '', text) #removendo @
    text = re.sub('#', '', text) #removendo #
    text = re.sub('RT[\s]+', '', text) # Removendo RT
    text = re.sub('https?:\/\/\S+', '', text) # Removendo hyperlink
    text = re.sub('&amp','', text)# removendo marcação HTML de início
    return text
```

<ul>
<li> <b> POLARITY </b> é um valor contínuo que varia de -1.0 a 1.0,  <b>100% negativo = -1.0</b> e <b>100% positivo = 1.0 </b> </li>
<li> <b>SUBJECTIVITY </b> que também é um valor contínuo que varia de 0.0 a 1.0, <b>100% objetivo = 0.0 </b> e <b>100% subjetivo = 1.0</b> </li>
<li> <b> Neutro </b>    sentimento neutro = 0</li>  
</ul>

Lembrando que nem sempre o algoritmo consegue fazer a análise.

> bolsonaro_df['sentimento'] = bolsonaro_df['tweet_without_stopwords'].apply(capture_polaridade_pt)

<b> chama a função  capture_polaridade_pt(text)</b>

> bolsonaro_df['subjetividade'] = bolsonaro_df['tweet_without_stopwords'].apply(capture_subjetivida_pt)

<b> chama a função  capture_subjetivity_pt(text)</b> 

```python
import re
import tweepy
from datetime import datetime
from textblob import TextBlob
```

```python
# Função que analisa e obtém a polaridade
#-1 negativo
#0 neutro
#1 positivo

def capture_polaridade_pt(text):
    frase = TextBlob(text)
    if (frase.detect_language() != 'en'):
        traducao = TextBlob(str(frase.translate(to='en')))
        return (traducao.sentiment.polarity)
    else:
        return (frase.sentiment.polarity)

# Função que analisa e obtém a subjetividade
def capture_subjetividade_pt(text):
    frase = TextBlob(text)
    if (frase.detect_language() != 'en'):
        traducao = TextBlob(str(frase.translate(to='en')))
        return (traducao.sentiment.subjectivity)
    else:
        return (frase.sentiment.subjectivity)
```

```python
# Mostra um novo dataframe com as colunas subjetividade e polaridade

bolsonaro_df['sentimento'] = bolsonaro_df['tweet_without_stopwords'].apply(capture_polaridade_pt)
bolsonaro_df['subjetividade'] = bolsonaro_df['tweet_without_stopwords'].apply(capture_subjetividade_pt)

mourao_df['sentimento'] = mourao_df['tweet_without_stopwords'].apply(capture_polaridade_pt)
mourao_df['subjetividade'] = mourao_df['tweet_without_stopwords'].apply(capture_subjetividade_pt)

minSaude1Mandeta_df['sentimento'] = minSaude1Mandeta_df['tweet_without_stopwords'].apply(capture_polaridade_pt)
minSaude1Mandeta_df['subjetividade'] = minSaude1Mandeta_df['tweet_without_stopwords'].apply(capture_subjetividade_pt)

minSaude2Teich_df['sentimento'] = minSaude2Teich_df['tweet_without_stopwords'].apply(capture_polaridade_pt)
minSaude2Teich_df['subjetividade'] = minSaude2Teich_df['tweet_without_stopwords'].apply(capture_subjetividade_pt)

minSaude4Queiroga_df['sentimento'] = minSaude4Queiroga_df['tweet_without_stopwords'].apply(capture_polaridade_pt)
minSaude4Queiroga_df['subjetividade'] = minSaude4Queiroga_df['tweet_without_stopwords'].apply(capture_subjetividade_pt)
```
